```python
import os
from tqdm import tqdm
```

```python
msa_dir = "/Users/akshayonly/Work/Sequence-Data/Subunits-Sequences/MSA"
profile_dir = "/Users/akshayonly/Work/Sequence-Data/Subunits-Sequences/Profiles/"

if not os.path.exists(profile_dir):
    os.makedirs(profile_dir)
```

```python
msa_files = sorted(os.listdir(msa_dir))

for msa in tqdm(msa_files):
    if msa.endswith(".fasta"):
        input_file = os.path.join(msa_dir, msa)

        profile = msa.replace("mafft_msa.fasta", "profile.hmm")
        output_file = os.path.join(profile_dir, profile)

        hmmbuild_command = f"hmmbuild --amino --plaplace {output_file} {input_file}"

        os.system(hmmbuild_command)
```
